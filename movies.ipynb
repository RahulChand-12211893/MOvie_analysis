```python
def numerical_univariate(df, col):
    print("Output of describe function")
    print(df[col].describe())
    print("Output of distribution plot")
    sns.distplot(df[col])
    plt.show()
    print("Output of boxplot")
    sns.boxplot(df[col])
    plt.show()
```

```python
def remove_outliers(df, col):

  # 25th percentile or 1st quartile
  Q1 = df[col].quantile(0.25)

  # 75th percentile or 3rd quartile
  Q3 = df[col].quantile(0.75)

  # IQR
  IQR = Q3-Q1

  # lower whisker or lower threshold
  lower_threshold = Q1 - 1.5*IQR

  # Upper whisker
  upper_threshold = Q3 + 1.5*IQR

  # remove all values less than lower threshold and greater than upper threshold
  return df[(df[col] > lower_threshold) & (df[col] < upper_threshold)]
```

```python
df_without_outliers = remove_outliers(df, 'votes')
```

```python
# Question23. Which movies have the longest and shortest runtimes, and how do their ratings compare?
if 'movie_title' in df.columns and 'runtime' in df.columns and 'imdb_rating' in df.columns:
    df_runtime = df.dropna(subset=['movie_title', 'runtime', 'imdb_rating'])
    longest_movie = df_runtime.loc[df_runtime['runtime'].idxmax()]
    shortest_movie = df_runtime.loc[df_runtime['runtime'].idxmin()]
    longest_movie_title = longest_movie['movie_title']
    longest_movie_runtime = longest_movie['runtime']
    longest_movie_rating = longest_movie['imdb_rating']
    shortest_movie_title = shortest_movie['movie_title']
    shortest_movie_runtime = shortest_movie['runtime']
    shortest_movie_rating = shortest_movie['imdb_rating']
    print(f"Longest Runtime Movie: {longest_movie_title} ({longest_movie_runtime} minutes) - IMDb Rating: {longest_movie_rating}")
    print(f"Shortest Runtime Movie: {shortest_movie_title} ({shortest_movie_runtime} minutes) - IMDb Rating: {shortest_movie_rating}")
    plt.figure(figsize=(10, 6))
    plt.bar(['Longest Runtime', 'Shortest Runtime'], [longest_movie_runtime, shortest_movie_runtime], color=['skyblue', 'lightcoral'])
    plt.ylabel('Runtime (minutes)')
    plt.title('Longest and Shortest Runtime Movies')
    plt.text(0, longest_movie_runtime + 5, f"IMDb Rating: {longest_movie_rating}", ha='center', fontsize=12)
    plt.text(1, shortest_movie_runtime + 5, f"IMDb Rating: {shortest_movie_rating}", ha='center', fontsize=12)
    plt.show()
else:
    print("The 'movie_title', 'runtime', or 'imdb_rating' column is not available in the dataset.")
```
